```python
df.drop(columns=['Lead_Investor_2'], inplace=True)
```

```python
df = df.dropna(subset=['Startup_Name'])  # drop rows with no name
df['Sector'] = df['Sector'].fillna('Unknown')
df['Location_City'] = df['Location_City'].fillna('Unknown')
df['Lead_Investor'] = df['Lead_Investor'].fillna('Unknown')
df['Outcome'] = df['Outcome'].fillna('Unknown')
df.isnull().sum()
```

```python
# Clean funding amount
df['funding_amt_clean_me'] = df['funding_amt_clean_me'].astype(str).str.replace(r'[\$,]', '', regex=True)
df['funding_amt_clean_me'] = pd.to_numeric(df['funding_amt_clean_me'], errors='coerce')

# Clean valuation
df['Valuation_Raw'] = df['Valuation_Raw'].astype(str).str.replace(r'[\$,]', '', regex=True)
df['Valuation_Raw'] = pd.to_numeric(df['Valuation_Raw'], errors='coerce')
df.isnull().sum()
```

```python
df['Funding_Date_Raw'] = pd.to_datetime(df['Funding_Date_Raw'], errors='coerce')
df.isnull().sum()
df.info()

```

```python
# Instead of using inplace, assign back to the column
df['Funding_Amount_Raw'] = df['Funding_Amount_Raw'].fillna(df['Funding_Amount_Raw'].median())
df.isnull().sum()
```

```python
df['funding_amt_clean_me'] = df['Funding_Amount_Raw'].astype(str).str.replace(r'[\$,]', '', regex=True)
df['funding_amt_clean_me'] = pd.to_numeric(df['funding_amt_clean_me'], errors='coerce')
```

```python
df = df.dropna(subset=['Funding_Date_Raw'])
```

```python
df.info()
```

```python
# Create a new column 'Year' from the Funding_Date_Raw
df['Year'] = df['Funding_Date_Raw'].dt.year

# Check the first few rows
df[['Startup_Name', 'Funding_Date_Raw', 'Year']].head()
```

```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# For better plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10,6)
```

```python
# Count number of startups per sector
top_sectors = df['Sector'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_sectors.values, y=top_sectors.index, palette='viridis')
plt.title("Top 10 Sectors by Number of Startups")
plt.xlabel("Number of Startups")
plt.ylabel("Sector")
plt.show()
```

```python
# Sum of funding per sector
funding_by_sector = df.groupby('Sector')['Funding_Amount_Raw'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=funding_by_sector.values, y=funding_by_sector.index, palette='magma')
plt.title("Top 10 Sectors by Total Funding")
plt.xlabel("Total Funding (INR)")
plt.ylabel("Sector")
plt.show()
```

```python
# Total funding per year
funding_by_year = df.groupby('Year')['Funding_Amount_Raw'].sum()

# Plot
plt.figure(figsize=(12,6))
sns.lineplot(x=funding_by_year.index, y=funding_by_year.values, marker='o')
plt.title("Total Startup Funding Over Years")
plt.xlabel("Year")
plt.ylabel("Total Funding (INR)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Funding_Amount_Raw'], bins=50, kde=True)
plt.xscale('log')  # Use log scale because funding varies a lot
plt.title("Distribution of Funding Amounts (Log Scale)")
plt.xlabel("Funding Amount (INR, log scale)")
plt.ylabel("Number of Startups")
plt.show()
```

```python
# Count number of investments per lead investor
top_investors = df['Lead_Investor'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_investors.values, y=top_investors.index, palette='coolwarm')
plt.title("Top 10 Lead Investors")
plt.xlabel("Number of Investments")
plt.ylabel("Investor")
plt.show()

```

```python
# Sum funding per city
funding_by_city = df.groupby('Location_City')['Funding_Amount_Raw'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=funding_by_city.values, y=funding_by_city.index, palette='cubehelix')
plt.title("Top 10 Cities by Total Funding")
plt.xlabel("Total Funding (INR)")
plt.ylabel("City")
plt.show()
```

```python
# Count of startups per funding round
funding_round_counts = df['Funding_Round'].value_counts()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=funding_round_counts.values, y=funding_round_counts.index, palette='Set2')
plt.title("Number of Startups by Funding Round")
plt.xlabel("Number of Startups")
plt.ylabel("Funding Round")
plt.show()
```

```python
# Create pivot table: sectors vs funding rounds, count of startups
sector_round_pivot = df.pivot_table(index='Sector', columns='Funding_Round', values='Startup_Name', aggfunc='count', fill_value=0)

# Plot heatmap
plt.figure(figsize=(12,8))
sns.heatmap(sector_round_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title("Number of Startups by Sector and Funding Round")
plt.xlabel("Funding Round")
plt.ylabel("Sector")
plt.show()
```

```python
# Average funding per sector
avg_funding_sector = df.groupby('Sector')['Funding_Amount_Raw'].mean().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=avg_funding_sector.values, y=avg_funding_sector.index, palette='rocket')
plt.title("Top 10 Sectors by Average Funding per Startup")
plt.xlabel("Average Funding (INR)")
plt.ylabel("Sector")
plt.show()
```

```python
# Create success indicator (example: Outcome = 'Successful' counts as 1)
df['Success'] = df['Outcome'].apply(lambda x: 1 if 'Successful' in x else 0)

# Success rate per sector
success_rate_sector = df.groupby('Sector')['Success'].mean().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=success_rate_sector.values, y=success_rate_sector.index, palette='Blues_r')
plt.title("Top 10 Sectors by Startup Success Rate")
plt.xlabel("Success Rate")
plt.ylabel("Sector")
plt.show()
```

```python
# Pivot table: count of startups by city and sector
city_sector_pivot = df.pivot_table(index='Location_City', columns='Sector', values='Startup_Name', aggfunc='count', fill_value=0)

# Optional: show top 10 cities with most startups overall
top_cities = df['Location_City'].value_counts().head(10).index
city_sector_top = city_sector_pivot.loc[top_cities]

# Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(city_sector_top, annot=True, fmt='d', cmap='coolwarm')
plt.title("Number of Startups by City and Sector (Top 10 Cities)")
plt.xlabel("Sector")
plt.ylabel("City")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (16, 20)

# Prepare data
top_sectors = df['Sector'].value_counts().head(10)
funding_by_sector = df.groupby('Sector')['Funding_Amount_Raw'].sum().sort_values(ascending=False).head(10)
funding_by_year = df.groupby('Year')['Funding_Amount_Raw'].sum()
top_investors = df['Lead_Investor'].value_counts().head(10)
top_cities = df['Location_City'].value_counts().head(10)
avg_funding_sector = df.groupby('Sector')['Funding_Amount_Raw'].mean().sort_values(ascending=False).head(10)
success_rate_sector = df.groupby('Sector')['Success'].mean().sort_values(ascending=False).head(10)

# Create subplots
fig, axes = plt.subplots(4, 2, figsize=(18, 24))
fig.tight_layout(pad=6.0)

# Plot 1: Top sectors by startup count
sns.barplot(x=top_sectors.values, y=top_sectors.index, palette='viridis', ax=axes[0,0])
axes[0,0].set_title("Top 10 Sectors by Number of Startups")
axes[0,0].set_xlabel("Number of Startups")
axes[0,0].set_ylabel("Sector")

# Plot 2: Top sectors by total funding
sns.barplot(x=funding_by_sector.values, y=funding_by_sector.index, palette='magma', ax=axes[0,1])
axes[0,1].set_title("Top 10 Sectors by Total Funding")
axes[0,1].set_xlabel("Total Funding (INR)")
axes[0,1].set_ylabel("Sector")

# Plot 3: Funding trend over years
sns.lineplot(x=funding_by_year.index, y=funding_by_year.values, marker='o', ax=axes[1,0])
axes[1,0].set_title("Total Startup Funding Over Years")
axes[1,0].set_xlabel("Year")
axes[1,0].set_ylabel("Total Funding (INR)")

# Plot 4: Top lead investors
sns.barplot(x=top_investors.values, y=top_investors.index, palette='coolwarm', ax=axes[1,1])
axes[1,1].set_title("Top 10 Lead Investors")
axes[1,1].set_xlabel("Number of Investments")
axes[1,1].set_ylabel("Investor")

# Plot 5: Top cities by startup count
sns.barplot(x=top_cities.values, y=top_cities.index, palette='cubehelix', ax=axes[2,0])
axes[2,0].set_title("Top 10 Cities by Number of Startups")
axes[2,0].set_xlabel("Number of Startups")
axes[2,0].set_ylabel("City")

# Plot 6: Average funding per sector
sns.barplot(x=avg_funding_sector.values, y=avg_funding_sector.index, palette='rocket', ax=axes[2,1])
axes[2,1].set_title("Top 10 Sectors by Average Funding per Startup")
axes[2,1].set_xlabel("Average Funding (INR)")
axes[2,1].set_ylabel("Sector")

# Plot 7: Success rate per sector
sns.barplot(x=success_rate_sector.values, y=success_rate_sector.index, palette='Blues_r', ax=axes[3,0])
axes[3,0].set_title("Top 10 Sectors by Startup Success Rate")
axes[3,0].set_xlabel("Success Rate")
axes[3,0].set_ylabel("Sector")

# Plot 8: Funding amount distribution (log scale)
sns.histplot(df['Funding_Amount_Raw'], bins=50, kde=True, ax=axes[3,1])
axes[3,1].set_xscale('log')
axes[3,1].set_title("Funding Amount Distribution (Log Scale)")
axes[3,1].set_xlabel("Funding Amount (INR, log scale)")
axes[3,1].set_ylabel("Number of Startups")

plt.show()
```
